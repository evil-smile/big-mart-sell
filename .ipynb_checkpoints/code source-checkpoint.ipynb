# Traitement de notre dataset pour l'ntrainement

```python
# traitement de la variable 'Item_Fat_Content'
df['Item_Fat_Content'].replace({'LF':'Low Fat', 'reg':'Regular', 'low fat':'Low Fat'}, inplace=True)
df.Item_Fat_Content.value_counts()
```

```python
# les valeurs manquantes
df.isna().sum()
```

 Nous allons essayer de bien comprendre les variables contenant les valeurs manquantes avant de passer au traitement de valeurs manquantes.

```python
sns.heatmap(df.isna())
```

Ce graphique nous laisse comprendre qu'il n'existe aucune relation sur le NaN entre les deux variables, donc si nous essayons de suprimer les cases contenant les valeurs manquantes, nous allons perdre pres de la moitier de nos donnees, donc il faut proceder autrement.

```python
# regardons la description de la variable categoriques
df.describe(include='object')
```

```python
df.Outlet_Size.value_counts()
```

```python
#df.Outlet_Size.fillna(method='bfill', inplace=True)
df.Outlet_Size.fillna(method='ffill', inplace=True)
df.Outlet_Size.value_counts()
```

```python
df.Outlet_Size.isna().sum()
```

**NB:** Nous avons decide d'utliser la methode de 'ffill' parce qu'on ne pouvait pas se base sur la description que nous offre le tableau de description, prendre le risque et remplacer les valeurs manquantes d'une variable categorique par une seule valeur peut entrainner un probleme de desequilibre de classe et lors de l'entrainement de notre modele, la machine risque de donner de tres mauvais score. 

```python
#la description de la variable numerique
df.describe()
```

```python
df.Item_Weight.fillna(value=df.Item_Weight.mean(), inplace=True)
df.isna().sum()
```

**NB:** Nous avons remplace les valeurs maquantes de cette variable par la moyenne de ses valeurs car il s'agit d'une variable numerique, car sont ecart-type est trop petit pour etre remplace a la place des NaN. Nous avons egalement tenu compte de la description de la variable

# Preparation de donnees et entrainement de notre modele

```python
#importation des donnees de teste
df_test = pd.read_csv('test_AbJTz2l.csv')
df_test.isna().sum()
```

```python
# je traite les valeurs manquantes
df_test.Item_Weight.fillna(value=df_test.Item_Weight.mean(), inplace=True)
df_test.Outlet_Size.fillna(method='ffill', inplace=True)
df_test.isna().sum()
```

**NB**: le fait de melanger les donner d'entrainement et les donnees de teste n'est pas autorise par les regles de ce concours, mais je le fais juste pour un but lucratif, puisse que ce projet est open source

```python
# preparation de nos donnees
y_train = df.Item_Outlet_Sales
x_train = df.drop(['Item_Outlet_Sales'], axis=1)
x_test = df_test
```

```python
# phase d'encodage de nos donnees
from sklearn.preprocessing import OrdinalEncoder
x_train = OrdinalEncoder().fit_transform(x_train)
x_test = OrdinalEncoder().fit_transform(x_test)
```

**NB:** ce type d'encodage n'est pas tres conseille pour les datasets dons la valeurs des variables categoriques n'est pas binaire, genre vrai ou faux, oui ou non,... Comme par exemple la variable "Item_Identifier" qui contient plus de 1000 valeurs differentes, l'encodage OrdinalEncoder() va attribuer a chacun une valeur numerique, donc la valeurs OUT12 aura 1 comme identifient, OUT13 aura 2 comme identifient,... donc s'il y a 1500 valeurs dans ce variable categorique, nous auront les chiffre allant de 1 a 1500 dans notre tableau. C'est la raison pour laquelle cela risque tres mal se comporter lors de l'entrainement de notre modele.

```python
# importation des bibliotheques d'entrainement
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import LinearSVR
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.linear_model import SGDRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
# entrainement du modele
liste_modeles = [KNeighborsRegressor(n_neighbors=10, n_jobs=3), LinearRegression(),LinearSVR(random_state=500, C=10),
                RandomForestRegressor(), BaggingRegressor(random_state=400), SGDRegressor(shuffle=True), 
                 DecisionTreeRegressor(random_state=500)]
for modeles in liste_modeles:
    model = make_pipeline(StandardScaler(), modeles)
    model.fit(x_train, y_train)
    print(f"{modeles.__class__.__name__:-<30} {model.score(x_train, y_train)}")
```

Nous choisissons le modele de foret aleatoire car c'est lui qui donne un score normale

```python
modele_rf = make_pipeline(StandardScaler(), RandomForestRegressor())
modele_rf.fit(x_train, y_train)
modele_rf.score(x_train, y_train)
```

```python
# Nous allons maintenant voir la prediction sur les donnees futures
y_pred = modele_rf.predict(x_test)
```

```python
# evaluation de notre modele avec l'erreur quadratique moyenne
from sklearn.metrics import *
print("MSE:", mean_squared_error(y_train[0:5681], y_pred))
```

```python
# la courbe d'apprentissage de notre modele
from sklearn.model_selection import learning_curve
N, train, val = learning_curve(modele_rf, x_train, y_train, 
                              train_sizes=np.linspace(0.1, 1, 50), cv=5)
```

```python
plt.plot(N, train.mean(axis=1), label='Entrainement')
plt.plot(N, val.mean(axis=1), label='validation')
plt.legend()
```
